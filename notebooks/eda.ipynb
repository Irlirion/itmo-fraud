```python
import pandas as pd

historical_data = pd.read_parquet("../data/historical_currency_exchange.parquet")
historical_data["date"] = pd.to_datetime(historical_data["date"])
historical_data
```

```python
fraud_data = pd.read_parquet("../data/transaction_fraud_data.parquet")
fraud_data['timestamp'] = pd.to_datetime(fraud_data['timestamp'])
fraud_data['date'] = pd.to_datetime(fraud_data['timestamp'].dt.date)
hist_long = (
    historical_data
    .melt(id_vars=['date'], var_name='currency', value_name='rate')
)
fraud_data = fraud_data.merge(
    hist_long,
    on=['date', 'currency'],
    how='left'
)
fraud_data['amount_usd'] = fraud_data['amount'] / fraud_data['rate']
fraud_data
```

```python
bool_cols = [
    "is_card_present",
    "is_outside_home_country",
    "is_high_risk_vendor",
    "is_weekend",
    "is_fraud",
]
fraud_data[bool_cols] = fraud_data[bool_cols].astype(bool)
```

```python
fraud_data.info()
```

```python
ts = pd.to_datetime(fraud_data["timestamp"], errors="coerce")
fraud_data["hour"] = ts.dt.hour
fraud_data["dow"] = ts.dt.dayofweek  # 0=Mon
fraud_data["month"] = ts.dt.month
fraud_data["is_night"] = fraud_data["hour"].isin([0, 1, 2, 3, 4, 5]).astype(int)
```

```python
import numpy as np

fraud_data["amount_log1p"] = np.log1p(fraud_data["amount_usd"].clip(lower=0))
```

```python
fraud_data["city_unknown"] = (
    fraud_data["city"].str.contains("Unknown", case=False, na=False).astype(int)
)
fraud_data["vendor_unknown"] = fraud_data["vendor"].isna().astype(int)
fraud_data["currency"] = fraud_data["currency"].fillna("UNK")
fraud_data["vendor_category"] = fraud_data["vendor_category"].fillna("UNK")
fraud_data["vendor_type"] = fraud_data["vendor_type"].fillna("UNK")
fraud_data["device"] = fraud_data["device"].fillna("UNK")
fraud_data["channel"] = fraud_data["channel"].fillna("UNK")
fraud_data["card_type"] = fraud_data["card_type"].fillna("UNK")
fraud_data["city_size"] = fraud_data["city_size"].fillna("UNK")
fraud_data["country"] = fraud_data["country"].fillna("UNK")
```

```python
low_card_cats = [
    "vendor_category",
    "vendor_type",
    "card_type",
    "device",
    "channel",
    "city_size",
    "currency",
]
K = 20
for c in low_card_cats:
    topk = fraud_data[c].value_counts().nlargest(K).index
    fraud_data[c] = np.where(fraud_data[c].isin(topk), fraud_data[c], "OTHER")
```

```python
bin_cols = [
    "is_card_present",
    "is_outside_home_country",
    "is_high_risk_vendor",
    "is_weekend",
    "is_night",
    "city_unknown",
    "vendor_unknown",
]
for c in bin_cols:
    fraud_data[c] = fraud_data[c].astype(int)
```

```python
num_cols = ["amount_log1p", "hour", "dow", "month"]
cat_cols = low_card_cats + []
bin_cols = bin_cols + []
```

```python
X_num = fraud_data[num_cols].astype("float32")
```

```python
from pandas import DataFrame


def one_hot_simple(series):
    # explicit, to avoid hidden placeholders
    vals = series.astype("category")
    mat = pd.get_dummies(vals, prefix=series.name, dtype=np.uint8)
    return mat


X_cat_list = [one_hot_simple(fraud_data[c]) for c in cat_cols]
X_cat = (
    pd.concat(X_cat_list, axis=1).astype("uint8")
    if X_cat_list
    else pd.DataFrame(index=fraud_data.index)
)

X_bin = fraud_data[bin_cols].astype("uint8")

# Concatenate final matrix
X: DataFrame = pd.concat([X_num, X_bin, X_cat], axis=1)
y = fraud_data["is_fraud"].astype(int)  # only for validation/insight
```

```python
from pyod.models.ecod import ECOD
from pyod.models.iforest import IForest
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm

X_model = X.astype("float32").values

scaler = StandardScaler(with_mean=True, with_std=True)
X_scaled = scaler.fit_transform(X_model)

ecod = ECOD(n_jobs=-1)  # fast, good baselines
ifo = IForest(n_estimators=200, max_samples="auto", random_state=42, n_jobs=-1)

models = [
    ecod,
    ifo,
]
for m in tqdm(models):
    m.fit(X_scaled)
```

```python
# Collect raw scores and z-score standardize each model’s output before ensembling
def zscore(a):
    a = np.asarray(a, dtype=np.float64)
    return (a - np.nanmean(a)) / (np.nanstd(a) + 1e-12)


scores = np.column_stack([zscore(m.decision_scores_) for m in models])
score_avg = scores.mean(axis=1)
score_max = scores.max(axis=1)

# Attach back to frame
fraud_data["anomaly_score_avg"] = score_avg.astype("float32")
fraud_data["anomaly_score_max"] = score_max.astype("float32")
```

```python
fraud_data
```

```python
from sklearn.metrics import precision_recall_fscore_support, average_precision_score
from pprint import pprint

fraud_rate = float(y.mean())  # e.g., 0.002 (0.2%) — compute from your data
quantile = 1.0 - fraud_rate if fraud_rate > 0 and fraud_rate < 1 else 0.995  # fallback
th = fraud_data["anomaly_score_avg"].quantile(quantile)

fraud_data["anomaly_flag"] = (fraud_data["anomaly_score_avg"] >= th).astype(int)

precision, recall, f1, _ = precision_recall_fscore_support(
    fraud_data["is_fraud"],
    fraud_data["anomaly_flag"],
    average="binary",
    zero_division=0,
)
ap = average_precision_score(fraud_data["is_fraud"], fraud_data["anomaly_score_avg"])

metrics = {
    "fraud_rate": fraud_rate,
    "anomaly_rate": float(fraud_data["anomaly_flag"].mean()),
    "precision_at_calibrated_threshold": float(precision),
    "recall_at_calibrated_threshold": float(recall),
    "f1_at_calibrated_threshold": float(f1),
    "avg_precision_PR_AUC": float(ap),
}
pprint(metrics)
```

**Ключевые выводы:**

- Уровень аномалий (≈19.97 %) совпадает с фактическим уровнем мошенничества, что свидетельствует об адекватной калибровке порога.  
- Precision и Recall на выбранном пороге равны ≈0.62, что указывает на сбалансированную детекцию мошеннических транзакций.  
- Среднее значение Precision-Recall AUC ≈0.66 демонстрирует хорошую способность модели отличать нормальные операции от аномалий.  

```python
# Helper: summarize by slice with clear KPIs
def slice_report(
    frame,
    by,
    score="anomaly_score_avg",
    flag="anomaly_flag",
    label="is_fraud",
    min_tx=500,
):
    g = (
        frame.groupby(by)
        .agg(
            tx=("transaction_id", "count"),
            anomaly_rate=(flag, "mean"),
            fraud_rate=(label, "mean"),
            score_p95=(score, lambda s: np.quantile(s, 0.95)),
            score_mean=(score, "mean"),
            amount_mean=("amount_usd", "mean"),
            amount_p95=("amount_usd", lambda s: np.quantile(s, 0.95)),
            outside_home_rate=("is_outside_home_country", "mean"),
            high_risk_vendor_rate=("is_high_risk_vendor", "mean"),
            card_present_rate=("is_card_present", "mean"),
            weekend_rate=("is_weekend", "mean"),
        )
        .reset_index()
    )
    g = g[g["tx"] >= min_tx].copy()
    baseline_anom = frame[flag].mean()
    baseline_fraud = frame[label].mean()
    g["anomaly_lift"] = g["anomaly_rate"] / max(baseline_anom, 1e-9)
    g["fraud_lift"] = (
        g["fraud_rate"] / max(baseline_fraud, 1e-9) if baseline_fraud > 0 else np.nan
    )
    g = g.sort_values(
        ["anomaly_lift", "score_p95", "tx"], ascending=[False, False, False]
    )
    return g


by_lists = {
    "vendor_category": ["vendor_category"],
    "channel": ["channel"],
    "device": ["device"],
    "country": ["country"],
    "card_type": ["card_type"],
    "hour_of_day": ["hour"],
    "vendor_type": ["vendor_type"],
    "city_size": ["city_size"],
    "country_vendorcat": ["country", "vendor_category"],
}

reports = {
    name: slice_report(fraud_data, by, min_tx=1000) for name, by in by_lists.items()
}

# Top hotspots examples
for name, rep in reports.items():
    print(f"\n=== Top hotspots: {name} ===")
    print(rep.head(10).to_string(index=False))
```
